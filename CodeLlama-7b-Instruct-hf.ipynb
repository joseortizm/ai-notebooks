```python
# Función para generar código a partir de un prompt
def generate_code(prompt, max_length=100, num_return_sequences=1):
    response = pipeline(prompt, max_length=max_length, num_return_sequences=num_return_sequences)
    return response

# Ejemplo de uso
prompt = "python code to loop from 1 to 10 and print the numbers is:"  # Un ejemplo de prompt
result = generate_code(prompt)

# Imprime el resultado
for i, r in enumerate(result):
    print(f"Response {i+1}:\n{r['generated_text']}\n")
```

```python
from transformers import AutoTokenizer
import transformers
import torch

model = "../ai-models/CodeLlama-7b-Instruct-hf"
device = torch.device('mps')

tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map=device,
)
```

```python
sequences = pipeline(
    'python code to loop from 1 to 10 and print the numbers is:',
    do_sample=True,
    top_k=10,
    temperature=0.1,
    top_p=0.95,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=200,
)
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```
